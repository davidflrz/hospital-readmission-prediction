```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configuración
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# Ruta al archivo
data_path = Path(r"C:\Users\davfl\OneDrive\Desktop\Maestria en Analitica de Datos\Big Data\healthcare-readmission-project\data\raw\diabetic_data.csv")

# Cargar datos
df = pd.read_csv(data_path, na_values='?')

print(f"Shape: {df.shape}")
print(f"\nPrimeras filas:")
df.head()
```

```python
# Información del dataset
print("Información del Dataset:")
print("="*60)
df.info()

print("\n" + "="*60)
print("Estadísticas descriptivas:")
print("="*60)
df.describe()
```

```python
# Valores faltantes
print("Análisis de Valores Faltantes:")
print("="*60)

missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'columna': missing.index,
    'valores_faltantes': missing.values,
    'porcentaje': missing_pct.values
}).sort_values('porcentaje', ascending=False)

missing_df[missing_df['valores_faltantes'] > 0]
```

```python
# Análisis de readmisión
print("Distribución de Readmisión:")
print("="*60)
print(df['readmitted'].value_counts())
print("\nPorcentajes:")
print(df['readmitted'].value_counts(normalize=True) * 100)

# Visualización
plt.figure(figsize=(6, 3))
df['readmitted'].value_counts().plot(kind='bar')
plt.title('Distribución de Readmisión')
plt.xlabel('Readmisión')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Crear copia del dataframe
df_prep = df.copy()

# Crear target binario: 1 si readmisión <30 días, 0 en otro caso
df_prep['readmitted_30days'] = (df_prep['readmitted'] == '<30').astype(int)

print("Nueva distribución del target:")
print(df_prep['readmitted_30days'].value_counts())
print("\nPorcentajes:")
print(df_prep['readmitted_30days'].value_counts(normalize=True) * 100)

# Visualización
plt.figure(figsize=(6, 3))
df_prep['readmitted_30days'].value_counts().plot(kind='bar')
plt.title('Readmisión en 30 días (Target Binario)')
plt.xlabel('Readmisión <30 días')
plt.ylabel('Frecuencia')
plt.xticks([0, 1], ['No (0)', 'Sí (1)'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Listar todas las columnas
print("Columnas del dataset:")
print("="*60)
for i, col in enumerate(df_prep.columns, 1):
    print(f"{i:2d}. {col}")

# Identificar columnas con muchos valores faltantes
print("\n" + "="*60)
print("Columnas con >50% valores faltantes:")
high_missing = missing_df[missing_df['porcentaje'] > 50]
print(high_missing)
```

```python
# Columnas a eliminar (>80% valores faltantes)
columns_to_drop = ['weight', 'max_glu_serum', 'A1Cresult', 'payer_code', 'medical_specialty']

print("Columnas a eliminar:")
for col in columns_to_drop:
    missing_pct = (df_prep[col].isnull().sum() / len(df_prep)) * 100
    print(f"  - {col}: {missing_pct:.2f}% valores faltantes")

# Eliminar columnas
df_clean = df_prep.drop(columns=columns_to_drop)

print(f"\nShape original: {df_prep.shape}")
print(f"Shape después de eliminar columnas: {df_clean.shape}")
```

```python
# Eliminar columnas de identificación (no aportan a predicción)
id_columns = ['encounter_id', 'patient_nbr']

# Eliminar también la columna original 'readmitted' (ya tenemos readmitted_30days)
df_clean = df_clean.drop(columns=id_columns + ['readmitted'])

print(f"Shape después de eliminar IDs: {df_clean.shape}")
print(f"\nColumnas restantes: {len(df_clean.columns)}")
```

```python
# Ver tipos de datos
print("Tipos de datos:")
print("="*60)
print(df_clean.dtypes.value_counts())

print("\n" + "="*60)
print("Columnas por tipo:")
print("="*60)

print("\nNuméricas (int64):")
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(numeric_cols)

print("\nCategóricas (object):")
categorical_cols = df_clean.select_dtypes(include=['object']).columns.tolist()
print(categorical_cols)
```

```python
# Revisar cuántos valores únicos tiene cada variable categórica
print("Cardinalidad de variables categóricas:")
print("="*60)

for col in categorical_cols:
    n_unique = df_clean[col].nunique()
    print(f"{col:30s}: {n_unique:4d} valores únicos")
    if n_unique <= 10:
        print(f"  Valores: {df_clean[col].unique()[:10]}")
    print()
```

```python
def categorize_diagnosis(diag_code):
    """
    Agrupa códigos ICD-9 en categorías médicas principales
    """
    if pd.isnull(diag_code) or diag_code == '?':
        return 'Unknown'
    
    # Convertir a string y obtener primeros dígitos
    diag_str = str(diag_code)
    
    # Extraer primera parte del código
    if diag_str.startswith('V') or diag_str.startswith('E'):
        first_digit = diag_str[0]
    else:
        try:
            first_digit = int(float(diag_str.split('.')[0]))
        except:
            return 'Unknown'
    
    # Categorización según ICD-9
    if first_digit == 'V':
        return 'Supplementary'
    elif first_digit == 'E':
        return 'Injury'
    elif 390 <= first_digit <= 459 or first_digit == 785:
        return 'Circulatory'
    elif 460 <= first_digit <= 519 or first_digit == 786:
        return 'Respiratory'
    elif 520 <= first_digit <= 579 or first_digit == 787:
        return 'Digestive'
    elif first_digit == 250:
        return 'Diabetes'
    elif 800 <= first_digit <= 999:
        return 'Injury'
    elif 710 <= first_digit <= 739:
        return 'Musculoskeletal'
    elif 580 <= first_digit <= 629 or first_digit == 788:
        return 'Genitourinary'
    elif 140 <= first_digit <= 239:
        return 'Neoplasms'
    elif (1 <= first_digit <= 139) or (680 <= first_digit <= 709):
        return 'Infectious'
    else:
        return 'Other'

# Aplicar categorización
df_clean['diag_1_category'] = df_clean['diag_1'].apply(categorize_diagnosis)
df_clean['diag_2_category'] = df_clean['diag_2'].apply(categorize_diagnosis)
df_clean['diag_3_category'] = df_clean['diag_3'].apply(categorize_diagnosis)

# Eliminar columnas originales de diagnóstico
df_clean = df_clean.drop(columns=['diag_1', 'diag_2', 'diag_3'])

print("Categorías de diagnóstico creadas")
print("\nDistribución de diag_1_category:")
print(df_clean['diag_1_category'].value_counts())
```

```python
print("Dataset final limpio:")
print("="*60)
print(f"Shape: {df_clean.shape}")
print(f"\nColumnas totales: {len(df_clean.columns)}")

# Verificar valores faltantes restantes
missing_final = df_clean.isnull().sum()
missing_final = missing_final[missing_final > 0].sort_values(ascending=False)

print(f"\nColumnas con valores faltantes:")
if len(missing_final) > 0:
    print(missing_final)
else:
    print("No hay valores faltantes")

print(f"\nDistribución del target:")
print(df_clean['readmitted_30days'].value_counts())
```

```python
import json
from datetime import datetime

def create_patient_document(row):
    """
    Convierte una fila del DataFrame en un documento MongoDB estructurado
    """
    doc = {
        # Información demográfica
        "demographics": {
            "race": row['race'] if pd.notna(row['race']) else "Unknown",
            "gender": row['gender'],
            "age": row['age']
        },
        
        # Información de admisión
        "admission": {
            "type_id": int(row['admission_type_id']),
            "source_id": int(row['admission_source_id']),
            "discharge_disposition_id": int(row['discharge_disposition_id']),
            "time_in_hospital": int(row['time_in_hospital'])
        },
        
        # Procedimientos y medicamentos
        "clinical": {
            "num_lab_procedures": int(row['num_lab_procedures']),
            "num_procedures": int(row['num_procedures']),
            "num_medications": int(row['num_medications']),
            "number_diagnoses": int(row['number_diagnoses'])
        },
        
        # Historial de utilización
        "utilization": {
            "number_outpatient": int(row['number_outpatient']),
            "number_emergency": int(row['number_emergency']),
            "number_inpatient": int(row['number_inpatient'])
        },
        
        # Diagnósticos (categorizados)
        "diagnoses": {
            "primary": row['diag_1_category'],
            "secondary": row['diag_2_category'],
            "additional": row['diag_3_category']
        },
        
        # Medicamentos para diabetes (solo los más importantes)
        "diabetes_medications": {
            "metformin": row['metformin'],
            "insulin": row['insulin'],
            "glyburide": row['glyburide'],
            "glipizide": row['glipizide'],
            "glimepiride": row['glimepiride'],
            "pioglitazone": row['pioglitazone'],
            "rosiglitazone": row['rosiglitazone']
        },
        
        # Información de tratamiento
        "treatment": {
            "change_in_medication": row['change'],
            "diabetes_medication_prescribed": row['diabetesMed']
        },
        
        # Target variable
        "outcome": {
            "readmitted_30days": int(row['readmitted_30days']),
            "readmitted_30days_label": "Yes" if row['readmitted_30days'] == 1 else "No"
        }
    }
    
    return doc

# Crear lista de documentos
print("Convirtiendo datos a formato MongoDB...")

documents = []
for idx, row in df_clean.iterrows():
    doc = create_patient_document(row)
    documents.append(doc)
    
    if (idx + 1) % 10000 == 0:
        print(f"Procesados: {idx + 1}/{len(df_clean)} registros")

print(f"\nTotal de documentos creados: {len(documents)}")
```

```python
# Guardar documentos como JSON
json_path = Path(r"C:\Users\davfl\OneDrive\Desktop\Maestria en Analitica de Datos\Big Data\healthcare-readmission-project\data\processed\hospital_readmission.json")

with open(json_path, 'w') as f:
    for doc in documents:
        f.write(json.dumps(doc) + '\n')

print(f"\nArchivo JSON guardado:")
print(f"  Ruta: {json_path}")
print(f"  Tamaño: {json_path.stat().st_size / (1024*1024):.2f} MB")
print(f"  Documentos: {len(documents)}")
```
